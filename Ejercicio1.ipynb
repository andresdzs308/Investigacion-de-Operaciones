```python
import cvxpy as cp
import numpy as np

# Definir las variables de optimización x e y
x = cp.Variable()
y = cp.Variable()

# Definir la función objetivo
objective = cp.Minimize(3*x + 8*y)

# Definir las restricciones
constraints = [x + y >= 50, x - y <= 20, x >= 0, y >= 0]

# Crear el objeto problema
problem = cp.Problem(objective, constraints)

# Resolver el problema utilizando el solver ECOS
solution1 = problem.solve(solver=cp.ECOS)

# Resolver el problema utilizando el solver SCS
solution2 = problem.solve(solver=cp.SCS)

# Comparar las soluciones generadas por cada solver y calcular las diferencias
print("Solución 1 (ECOS): x = {}, y = {}, valor objetivo = {}".format(x.value, y.value, solution1))
print("Solución 2 (SCS): x = {}, y = {}, valor objetivo = {}".format(x.value, y.value, solution2))
print("Diferencia en el valor objetivo: {}".format(solution1 - solution2))
```
